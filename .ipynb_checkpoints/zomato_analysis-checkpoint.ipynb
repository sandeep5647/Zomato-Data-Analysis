# Zomato Data Analysis

This Jupyter Notebook contains data analysis and visualizations for a Zomato dataset. The analysis answers the following questions:

1. What type of restaurant do the majority of customers order from?
2. How many votes has each type of restaurant received?
3. What are the ratings that the majority of restaurants have received?
4. What is the average spending of couples per order?
5. Which mode (online or offline) has received the maximum rating?
6. Which type of restaurant received more offline orders?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset (replace 'zomato.csv' with the actual file path)
data = pd.read_csv('zomato_data.csv')
```

```python
# Convert 'rate' to numeric (handling 'x/5' format and non-numeric values)
data['rate'] = data['rate'].str.extract('(\\d+\\.\\d+)').astype(float)
```

```python
# Convert all values in the column to strings, then remove commas, and convert to numeric
data['approx_cost(for two people)'] = (
    data['approx_cost(for two people)']
    .astype(str)  # Convert to string
    .str.replace(',', '')  # Remove commas
    .replace('nan', None)  # Handle cases where values are 'nan' (as strings)
    .astype(float)  # Convert to numeric
)
```

# Analysis and Visualization

## 1. What type of restaurant do the majority of customers order from?

```python
restaurant_type_counts = data['listed_in(type)'].value_counts()
plt.figure(figsize=(8, 5))
restaurant_type_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Restaurants by Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

# 2. How many votes has each type of restaurant received?

```python
votes_by_type = data.groupby('listed_in(type)')['votes'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
votes_by_type.plot(kind='bar', color='coral')
plt.title('Total Votes by Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.show()
```

# 3. What are the ratings that the majority of restaurants have received?

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['rate'], bins=10, kde=True, color='green')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

# 4. What is the average spending of couples per order?

```python
average_spending = data['approx_cost(for two people)'].mean()
print(f"The average spending of couples per order is: {average_spending:.2f}")
```

# 5. Which mode (online or offline) has received the maximum rating?

```python
average_rating_by_mode = data.groupby('online_order')['rate'].mean()
average_rating_by_mode.plot(kind='bar', color=['orange', 'blue'], figsize=(8, 5))
plt.title('Average Rating by Order Mode')
plt.xlabel('Online Order')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.show()
```

# 6. Which type of restaurant received more offline orders?

```python
offline_orders = data[data['online_order'] == 'No'].groupby('listed_in(type)').size().sort_values(ascending=False)
offline_orders.plot(kind='bar', color='purple', figsize=(8, 5))
plt.title('Offline Orders by Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Number of Offline Orders')
plt.xticks(rotation=45)
plt.show()
```
